```python
def tiempo(seg):
  #encontrar el número de horas
  seg_restantes = seg
  try:
    if (seg_restantes >=3600):
      horas = int(seg_restantes/3600)
      seg_restantes = seg - 3600*horas
    else:
      horas =0
    
    if (seg_restantes >=60):
      minutos = int(seg_restantes/60)
      seg_restantes = seg_restantes - 60*minutos
    else:
      minutos =0

    print(f"Horas: {horas}, Minutos: {minutos}, Segundos: {seg_restantes}")


  except:
    print("Asegúrese de ingresar números enteros")
```

Testeando resultados. 

```python
try:
  time = int(input("Ingrese los segundos a convertir: "))
  tiempo(time)
  
except ValueError:
  print("Por favor ingresar un número entero, no texto o decimales")
```

```python
def eliminarTuplaVacia(lista):
   
   for indice in range(0,len(lista)-1):
     if len(lista[indice]) == 0:
       lista.pop(indice)
  
   return lista
```

#**Ejercicio 6 - Solución**

p6.JPG

```python
def promedios(tuplas):
  promedios = []
  for tupla in tuplas:
    suma =0
    for elemento in tupla:
      suma +=elemento
    promedio = suma/len(tupla) 
    promedios.append(promedio)
  
  return tuple(promedios)

```

```python
#Testeando
tupla = ((10,10,10,12),(30,45,56,45),(81,80,39,32))
promedios(tupla)
```

```python
from ast import Pass
facturas= {"No": None, "Costo": None}
numeros = []
costos = []

def actualizar():
  facturas["No"]=numeros
  facturas["Costo"]=costos

def add():

  numero= int(input("Ingrese el número de la factura: > "))
  numeros.append(numero)

  costo = input("Ingrese el valor  de la factura, en Q. > ")
  costos.append(costo)

#Los agregamos al diccionario.
  actualizar()
  show()


def pay():
  nofact= int(input("ingrese el número de factura que desea pagar: > "))
  index=0

  while index < len(numeros):
    if numeros[index]==nofact:
      numeros.pop(index)
      costos.pop(index)
      print(f"Factura {nofact} pagada exitosamente!")
      index = len(numeros)+5  #Hacemos que salga del loop
    elif index == len(numeros)-1: #Si llegamos al final sin encontrar nada, nos salimos.
      print("El número de factura no existee")
    index +=1



  actualizar()  #Actualizamos nuestro diccionario.
  show()


def show():
  no = "Numero de factura"
  val = "Valor de la factura"
  print("\n             Todas las facturas \n ")

  print(f"{no:<20} | {val: <20}  |")
  for i in range(len(numeros)):
      print(f"{numeros[i]:^20} | Q.{costos[i]: ^20}|")


menu = f"\nIngrese el número de lo que desea hacer:\n1. Agregar factura \n2. Pagar Factura  \n3. Terminar programa \n >"

choice=0
while choice !=3:
  choice = int(input(menu))
  if choice == 1:
    add()
  elif choice == 2:
    pay()

  elif choice ==3:
    print("Cerraremos el programa")
  else:
    print("Ingrese una opción valida.")
```

```python
def crearCartas():
  tipos = ["T","C","P","E"]
  numeros =[i for i in range(1,14)]
  lista = []
  #Creamos las cartas
  for tipo in tipos:
    for numero in numeros:
      carta = f"{tipo}{numero}"
      lista.append(carta)

  #Convertimos la lista a un set 
  cartas=set(lista)
  return cartas

def creandoSets():
  import random

  lista = list(crearCartas())

  n1 = random.randint(1,20)   #Limite el tamaño máximo a 20 cartas
  n2 = random.randint(1,20)

  mano1 = set(random.sample(lista, n1))
  mano2 = set(random.sample(lista, n2))

  return mano1,mano2

#-----------------------------------------------------------------
setA,setB=creandoSets()    #Cargamos 2 sets.
intercepción = setA.intersection(setB)
union = setA.union(setB)
print("Cartas:", crearCartas())
print("Conjunto A:", setA)
print("Conjunto B:", setB)
print("Intersepción A y B: ", intercepción)
print("Union A con B: ", union)

print()
print("¿Eventos independientes o no?")
print()
prob_intercepción= len(intercepción)/52  

prob_A = len(setA)/52
prob_B = len(setB)/52

prob_product = prob_A*prob_B

print("Probabilidad de Intercepción:",prob_intercepción )
print("Probilidad de A y B:",prob_product )

if prob_intercepción==0:
  print("Los eventos son independientes.")
else:
  print("Los eventos no son independientes.")

print()
print("Demostrando la regla de la adición")
print()

prob_union = len(union)/52

print("Probabilidad de la unión: ", prob_union)
prob_regla_suma= prob_A+prob_B-prob_intercepción
print("Probabilidad con la regla de la suma: ", prob_regla_suma)
print("Se puede ver que se obtienen los mismos resultados, por lo que queda demostrado la regla de la suma.")
```
